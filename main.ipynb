```python
from functions import (
    generate_data_serieses,
    generate_line_sereis,
    arrange_series,
    accumulate_series,
    transform_series,
    normalise_series,
    pad_series,
    normalise_factor,
)
import numpy as np
import math
from random import random
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
size = 1024
x=np.arange(size)

serieses = []
lines = []
for _ in range(20):
    left_size = math.floor(random() * 10) + 1
    right_size = math.floor(random() * 10) + 1
    middle_size = size - left_size - right_size

    pos_series, neg_series = generate_data_serieses(middle_size)
    series = arrange_series(pos_series, neg_series)
    series = pad_series(series, left_size, right_size)
    series = accumulate_series(series)
    m = (random() - 0.5) * 0.1
    c = (random() - 0.5) * 0.1

    line = generate_line_sereis(m, c, size)
    series = transform_series(series, line)
    
    nf = normalise_factor(series)
    serieses.append(normalise_series(series))
    lines.append((line - nf.min)/nf.factor)
    
```

```python
fig = make_subplots(4, 5)

for i in range(4):
    for j in range(5):
        fig.append_trace(go.Scatter(x=x, y=serieses[4*i + j]), row=i+1, col=j+1)
        fig.append_trace(go.Line(x=x, y=lines[4*i + j]), row=i+1, col=j+1)

fig.update_layout(height=600, width=800, showlegend=False)
fig.show()
```
