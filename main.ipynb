```python
from functions import generate_net_zero_series, accumulate_series
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go

size = 1024
x=np.arange(size)

serieses = []
for _ in range(20):
    serieses.append(accumulate_series(generate_net_zero_series(size)))


fig = make_subplots(4, 5)

for i in range(4):
    for j in range(i, 5):
        fig.append_trace(go.Scatter(x=x, y=serieses[i + j]), row=i+1, col=j+1)

fig.update_layout(height=800, width=1200, showlegend=False)
fig.show()
```
