```python
from functions import generate_data_serieses, arrange_series, accumulate_series
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
size = 1024
x=np.arange(size)

serieses = []
for _ in range(20):
    pos_series, neg_series = generate_data_serieses(size)
    series = arrange_series(pos_series, neg_series)
    series = accumulate_series(series)
    serieses.append(series)
len(series)
```

```python
fig = make_subplots(4, 5)

for i in range(4):
    for j in range(5):
        fig.append_trace(go.Scatter(x=x, y=serieses[4*i + j]), row=i+1, col=j+1)

fig.update_layout(height=600, width=800, showlegend=False)
fig.show()
```
